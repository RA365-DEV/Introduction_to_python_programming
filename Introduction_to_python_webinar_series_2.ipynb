**a) While Loop**

The while loop can be found in almost every programming languages and is utilized for iterative (or to repeat) one or more code **statements** for as long as the test condition (expression) is **True**. 

This iteration construct is usually used when the **number of times** we need to repeat the block of code to execute is **not known**.

image.png

The general form of a while loop;

image.png

```
count = 1 # Initialize counter
print("Initializing count")
while count < 8: # Should we continue?
print(count) # Display counter, then
count += 1 # Increment counter
```

### **do-while**

Python doesn't have a do-while. However, we can handle such by hard coding it.

```
a = 10
while True:
   a = a-1
  print(a)
  if a<7:
     break
print('Done.')
```

### Definite vs Indefinite Loop
When we know the exact number of iteration then, its a definite loop but when we have no idea the exact number of iteration, it becomes an indefinite loop. Let's look at the examples below: 

```
#Definite Loop
number = 1 #initialize varaible
print("Enter a number: ") #prompt for input
stop = int(input()) #Read input
while number <= stop: #iterate using condition
      print(number) #print output
      number += 1 #increase count
```

```
#Indefinite Loop
done = False # Enter the loop at least once
while not done:
      entry = int(input()) # Get value from user
      if entry == 100: # Did user provide the magic number?
        done = True # If so, get out
      else:
        print(entry) # If not, print it and continue
```

**b) For Loop**

**for loops** iterate over a collection of items, such as list or dict, and run a block of code with each element from
the collection. ***While loop*** can do the same thing as ***for loop***, but ***for loop*** is a more **concise** way fo doing what ***while loop*** can do.

image.png

image.png

```
num = 1 # Initialize sum
for i in range(1, 100, 10): #for loop with step size
    print(num)
```

```
sum = 0 # Initialize sum
for i in range(1, 100): #for loop with no step size
sum += i
print(sum)
```

```
#Use for loop to iterate and print the member of a list
for i in [0, 1, 2, 3, 4]: 
  print(i)
```

```
#Use for loop to iterate and print the member of a list
for x in ['one', 'two', 'three', 'four']:
  print(x)
```

```
#Nested For Loop
for i in range(1, 50, 5):
  print(i)
  for j in range(1, 100, 10):
    print(j)
```

```
# A practical Example
# Require the user to enter an integer in the range 1-10
value = 0 # Ensure loop entry
attempts = 0 # Count the number of tries
# Loop until the user supplies a valid number
while value >= 1 or value <= 5:
    value = int(input("Please enter an integer in the range 0-5: "))
    attempts += 1
    # value at this point is guaranteed to be within range
    if value >= 1 or value <=5
       print("It took you", attempts, "to enter a valid number. Great work.")
    else:
      print("For all", attempts, " trials, you enter didn't get it right. Shame on you.")
```

**c) Break and Continue**

image.png

```
# Only print code if all iterations completed 
num = int(input('Enter a number to check for: '))
for i in range(0, 6):
    if i == num:
        break
    print(i, ' ', end='')
print('Done')
```

```
#List bounds

a = [5, 10, 15, 20, 25]
print(a[7]) # Out-of-bounds access
```

```
#Slicing
list_built4 = [5,10,15,20,25,30,35,40,45,50,55,60,65,70]
print(list_built4[0:3]) # [5, 10, 15]
print(list_built4[4:8]) # [25, 30, 35, 40]
print(list_built4[2:5]) # [15, 20, 25]
print(list_built4[::-1]) #[70, 65, 60, 55, 50, 45, 40, 35, 30, 25, 20, 15, 10, 5]
print(list_built4[2:-2:2]) #[15, 25, 35, 45, 55]
print(list_built4[-5:-3]) # [50, 55]
```

**c) Building dictionary, slicing dictionary, modifying dictionary**

While lists and tuples are effective for storing collections of data, however they possess some limitations. How? we can
find an element within a list or tuple using their position (**index**). This approach can be utilized for many
applications, but in some situations access-by-index approach becomes inefficient.

A Python dictionary is an associative container which permits access based on a key, rather than an index.

Fundamental properties of Python Dictionary.

    1. Dictionary consists of key-value pairs.

    2. It is enclosed by curly braces {}.

    3. The stored values can be accessed using square brackets[].

```
dict1 = {'name' : 'red', 'age' : 10}
print(dict1) #will output all the key-value pairs. {'name':'red','age':10}
print(dict1['name']) #will output only value with 'name' key. 'red'
print(dict1.values()) #will output list of values in dic. ['red',10]
print(dict1.keys())
```

```
#Building dictionary
person = {}
type(person)
```

```
#Building dictionary
person['firstname'] = 'Joe'
person['othername'] = 'Itopa'
person['age'] = 51
person['maritalstatus'] = 'python'
person['children'] = ['Laptop', 'AI Algorithms', 'MLOps']
person['pets'] = {'dog' : 'ubuntu', 'cat': 'snow'}
#print(person)
print(person['maritalstatus'])
```

```
#Slicing Dictionary
data = {2:4, 3:9, 4:16, 5:25, 6:36, 7:49, 8:64}
interest = (2,7)
{key: data[key] for key in interest}
```

```
#Modify dictionary
animals = {'dog' : ['red','blue','green'], 'cat' : 'grey'}
animals['cat'] = 'red'
print(animals)
```
